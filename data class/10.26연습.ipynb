```python
#실습6 데이터에 피벗 사용하기
df = pd.read_csv('/home/piai/바탕화면/1 Big Data 분석 기초/FITNESS.csv',encoding='euc-kr')
```

```python
group_df = df.groupby(['GENDER','AGEGROUP']).agg('mean')
group_df
```
